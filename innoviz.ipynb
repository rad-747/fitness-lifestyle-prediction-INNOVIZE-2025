```python

# Fill missing values using random sampling from existing values
df.loc[df['diet_pref'].isnull(), 'diet_pref'] = np.random.choice(df['diet_pref'].dropna(), size=df['diet_pref'].isnull().sum(), replace=True)
```

```python

unique_categories = df['act_level'].dropna().unique()
probabilities = df['act_level'].dropna().value_counts(normalize=True).reindex(unique_categories, fill_value=0).values

# Ensure probabilities sum to 1 (avoid floating-point errors)
probabilities = probabilities / probabilities.sum()

# Fill missing values using weighted probabilities
df.loc[df['act_level'].isnull(), 'act_level'] = np.random.choice(unique_categories, size=df['act_level'].isnull().sum(), p=probabilities)
```

```python

# Fill missing values by randomly choosing from existing (non-null) values
df.loc[df['career'].isnull(), 'career'] = np.random.choice(df['career'].dropna(), size=df['career'].isnull().sum(), replace=True)
```

```python

# Fill missing value by randomly picking from existing values
df.loc[df['gender'].isnull(), 'gender'] = np.random.choice(df['gender'].dropna())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 1. Define Target Variable
y = df['is_healthy']

# 2. Define Features (Drop Target Column)
X = df.drop(columns=['is_healthy'])

# 3. One-Hot Encode Nominal Variables
X_encoded = pd.get_dummies(X, columns=['diet_pref', 'career'], drop_first=True)

# 4. Ordinal Encoding for Ordered Categories
act_level_order = [['Sedentary', 'Lightly Active', 'Active', 'Highly Active']]
encoder = OrdinalEncoder(categories=act_level_order)
X_encoded['act_level'] = encoder.fit_transform(X_encoded[['act_level']])

# 5. Scaling (Necessary for Logistic Regression)
numerical_cols = ['phy_fitness', 'daily_avg_steps', 'daily_avg_calories', 'sleep_hrs', 'mindfulness']

scaler = StandardScaler()
X_encoded[numerical_cols] = scaler.fit_transform(X_encoded[numerical_cols])

# 6. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

# 7. Train Logistic Regression Model
log_model = LogisticRegression()
log_model.fit(X_train, y_train)

# 8. Predictions
y_pred = log_model.predict(X_test)

# 9. Accuracy & Evaluation
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score

# Training Accuracy
train_acc = accuracy_score(y_train, log_model.predict(X_train))
print(f"Training Accuracy: {train_acc:.4f}")

# Testing Accuracy
test_acc = accuracy_score(y_test, log_model.predict(X_test))
print(f"Testing Accuracy: {test_acc:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

# --- Convert X_train (a DataFrame) to a NumPy array ---
X_train_np = X_train.to_numpy()

# --- Get feature names from the encoded DataFrame ---
feature_names = X_encoded.columns  # X_encoded has the final feature names after encoding and scaling
phy_idx = list(feature_names).index('phy_fitness')
mind_idx = list(feature_names).index('mindfulness')

# --- Create a mesh grid in the 2D space for phy_fitness and mindfulness ---
phy_min = X_train_np[:, phy_idx].min() - 0.5
phy_max = X_train_np[:, phy_idx].max() + 0.5
mind_min = X_train_np[:, mind_idx].min() - 0.5
mind_max = X_train_np[:, mind_idx].max() + 0.5

xx, yy = np.meshgrid(np.linspace(phy_min, phy_max, 200),
                     np.linspace(mind_min, mind_max, 200))

# --- Create full input vectors for each grid point ---
# Use the mean value of all features from X_train for features other than phy_fitness and mindfulness.
n_points = xx.ravel().shape[0]
mean_features = np.mean(X_train_np, axis=0)
grid_full = np.tile(mean_features, (n_points, 1))
grid_full[:, phy_idx] = xx.ravel()
grid_full[:, mind_idx] = yy.ravel()

# --- Predict using the full model (log_model trained on all features) ---
Z = log_model.predict(grid_full)
Z = Z.reshape(xx.shape)

# --- Plot the decision boundary and overlay the training points ---
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
sc = plt.scatter(X_train_np[:, phy_idx], X_train_np[:, mind_idx], c=y_train, cmap='coolwarm', edgecolor='k', s=60)
plt.xlabel("phy_fitness (scaled)")
plt.ylabel("mindfulness (scaled)")
plt.title("Decision Boundary using phy_fitness and mindfulness\n(Based on Full Model Trained on All Features)")

# --- Create a custom legend for the scatter plot with healthy/not healthy labels ---
unique_classes = np.unique(y_train)
handles = []
for cls in unique_classes:
    # Define the color based on the colormap. For binary classes, 
    # we can assign: 0 -> Not Healthy, 1 -> Healthy.
    norm_value = (cls - unique_classes.min()) / (unique_classes.max() - unique_classes.min() + 1e-8)
    color = plt.cm.coolwarm(norm_value)
    label = "0" if cls == 1 else "1"
    handles.append(mpatches.Patch(color=color, label=label))
plt.legend(handles=handles, title='Is Healthy', loc='best')

plt.show()
```
